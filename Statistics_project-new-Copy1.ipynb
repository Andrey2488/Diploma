* В качестве первого экономического показателя возьмем: число умерших на первом году жизни детей за год, по всем регионам, в сельской местности в 2019 году.

```python
def child_mortality_rural():
    path = ('social_russia_data/child_mortality_rural_1990_2021.xls')
    cur_df = pd.read_excel(path)
    name_col_ch_mort_rural = list(cur_df.iloc[1])
    name_col_ch_mort_rural[0] = 'Район'
    
    cur_df.columns = name_col_ch_mort_rural
    cur_df.drop(labels = [0, 1],axis = 0, inplace = True)
    
    cur_df = cur_df[['Район', 2019.0]]
    cur_df.rename(columns={2019.0: "Умерших детей 1 год чел, сельская(2019 год)"}, inplace=True)
    
    return cur_df
```

```python
cur_df = child_mortality_rural()
cur_df
```

* В качестве второго экономического показателя возьмем - число умерших на первом году жизни детей за год, по всем регионам, в городской местности за 2019 год.

```python
def child_mortality_urban(cur_df):
    path = ('social_russia_data/child_mortality_urban_1990_2021.xls')
    df = pd.read_excel(path)
    
    name_col_ch_mort_urban = list(df.iloc[1])
    name_col_ch_mort_urban[0] = 'Район'
    df.columns = name_col_ch_mort_urban
    df.drop(labels = [0, 1],axis = 0, inplace = True)
    
    cur_df['Умерших детей 1 год чел, городская(2019 год)'] = df[2019.0] # полностью совпадают регионы
    return cur_df
```

* В качестве третьего экономического показателя возьмем - число людей с инвалидностью по регионам, по месяцам, по возрастным группам за 2019 год. Месяцы буду усреднять по году.

Делаю преобразование столбцов так как таблицы с инвалидностью по регионам и общей агрегатной(полученной до текущего момента) отличались

```python
def predobr(cur_df, region):
    cur_df.loc[cur_df[region] == 'Республика Адыгея (Адыгея)', region] = 'Республика Адыгея'
    cur_df.loc[cur_df[region] == 'Кемеровская область - Кузбасс', region] = 'Кемеровская область'
    cur_df.loc[cur_df[region] == 'Ямало-Ненецкий автономный округ (Тюменская область)', region] = 'Ямало-Ненецкий автономный округ'
    cur_df.loc[cur_df[region] == 'Республика Татарстан (Татарстан)', region] = 'Республика Татарстан'
    cur_df.loc[cur_df[region] == 'Ханты-Мансийский автономный округ - Югра (Тюменская область)', region] = 'Ханты-Мансийский автономный округ'
    cur_df.loc[cur_df[region] == 'Чувашская Республика - Чувашия', region] = 'Чувашская Республика'
    cur_df.loc[cur_df[region] == 'Южный федеральный округ (с 29.07.2016)', region] = 'Южный федеральный округ'
    cur_df.loc[cur_df[region] == 'Город Москва столица Российской Федерации город федерального значения', region] = 'город Москва'
    cur_df.loc[cur_df[region] == 'Город Санкт-Петербург город федерального значения', region] = 'город Санкт-Петербург'
    cur_df.loc[cur_df[region] == 'Город федерального значения Севастополь', region] = 'город Севастополь'
    cur_df.loc[cur_df[region] == 'Ненецкий автономный округ (Архангельская область)', region] = 'Ненецкий автономный округ'
    return cur_df
```

```python
cur_df = predobr(cur_df)
```

```python
def disabled(cur_df):
    path2 = ('social_russia_data/disabled_total_by_age_2017_2022.csv')
    df2 = pd.read_csv(path2)
    
    cur_df['Район'] = cur_df['Район'].apply(lambda x: x.strip())
    region_cur = set(cur_df['Район'])
    region_1  = set(df2[df2['date'] == '2019-01-01']['region'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете poverty = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))
    cur_df['total_disabled'] = ''
    cur_df['18_30_disabled'] = ''
    cur_df['31_40_disabled'] = ''
    cur_df['41_50_disabled'] = ''
    cur_df['51_60_disabled'] = ''
    cur_df['60_disabled'] = ''
    
    for el in region_cur:
        for age in list(df2.columns[1:-1]):
            cur_df.loc[cur_df['Район'] == el, f'{age}_disabled'] = df2[df2['region'] == el][age].mean()

    return cur_df
```

* В качестве пятого экономического показателя возьмем - процент людей, живущих за чертой бедности (с денежными доходами ниже величины прожиточного минимума), оценка за год по регионам за 2019 год.

```python
def poverty_add_row(cur_df):
    
    path5 = ('social_russia_data/poverty_percent_by_regions_1992_2020.csv')
    df5 = pd.read_csv(path5)
    df5['region'] = df5['region'].apply(lambda x: x.strip())
    df5 = predobr(df5, 'region')
    region_cur = set(cur_df['Район'])
    region_1  = set(df5[df5['year'] == 2019]['region'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете poverty = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
    
    cur_df['poverty_percent'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'poverty_percent'] = list(df5[(df5['region'] == el) & (df5['year'] == 2019)]['poverty_percent'])[0]
        #print(el, int(df5[(df5['region'] == el) & (df5['year'] == 2019)]['poverty_percent']))
    
    return cur_df
```

* В качестве шестого экономического показателя возьмем - расходы на социальную политику от общих расходов бюджета региона, % в год

```python
def welfare(cur_df):
    path = ('social_russia_data/welfare_expense_share_2015_2020.xlsx')
    df = pd.read_excel(path)

    region_cur = set(cur_df['Район'])
    region_1  = set(df['region'])
    region_cur = region_cur.intersection(region_1)
    
    print("кол-во регионов в датасете welfare = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
    
    cur_df['welfare'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'welfare'] = list(df[df['region'] == el][2019.0])[0]
        
    return cur_df
```

Видим много \n, г. Сделаем обработку всех случаев, так как все округа встречаются в районе текущего датасета.

```python
def welfare_new(cur_df):
    path = ('social_russia_data/welfare_expense_share_2015_2020.xlsx')
    df = pd.read_excel(path)    
    for el in df['region']:
        if el.find("Республика") >= 0:
            el1 = el.replace("\n", " ")
        else:
            el1 = el.replace("\n", "")
        if el.find("г.") >= 0:
            el1 = el.replace("г.", "город")
        if el.find(" - Югра") >= 0:
            el1 = el.replace(" - Югра", "")
            el1 = el1.replace("\n", "")
        df.loc[df['region'] == el, 'region'] = el1
        
    region_cur = set(cur_df['Район'])
    region_1  = set(df['region'])
    region_cur = region_cur.intersection(region_1)
    
    print("кол-во регионов в датасете welfare = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
    
    cur_df['welfare'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'welfare'] = list(df[df['region'] == el][2019.0])[0]
        
    return cur_df
```

```python
cur_df = welfare_new(cur_df)
```

* В качестве седьмого экономического показателя возьмем: среднедушевые и реальные денежные доходы населения, номинальная и реальная начисленная зарплата, по регионам

```python
def cash_real_income(cur_df):
    path = ('social_russia_data/cash_real_income_wages_2015_2020.xlsx')
    # сделаю предобработку регионов, чтобы можно было джойнить друг с другом таблицы
    cur_df.loc[cur_df['Район'] == 'Архангельская область (кроме Ненецкого автономного округа)', 'Район'] = 'Архангельская область безавтономного округа'
    cur_df.loc[cur_df['Район'] ==  'Тюменская область (кроме Ханты-Мансийского автономного округа-Югры и Ямало-Ненецкого автономного округа)', 'Район'] = 'Тюменская область безавтономных округов'
    df = pd.read_excel(path)
    for el in df['region']:
        if el.find("Республика") >= 0:
            el1 = el.replace("\n", " ")
        else:
            el1 = el.replace("\n", "")
        if el.find("г.") >= 0:
            el1 = el.replace("г.", "город")
        if el.find(" - Югра") >= 0:
            el1 = el.replace(" - Югра", "")
            el1 = el1.replace("\n", "")
        df.loc[df['region'] == el, 'region'] = el1

    region_cur = set(cur_df['Район'])
    region_1  = set(df['region'])

    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете cash_real_income = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))

    cur_df['cash_real_income'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'cash_real_income'] = int(df[df['region'] == el][2019.0])
   
    return cur_df
```

```python
cur_df = cash_real_income(cur_df)
```

* В качестве восьмого экономического показателя возьмем: распределение малоимущего населения по социально-демографическим группам (дети, трудящиеся, пенсионеры) за 2017–2020 гг., по регионам.

```python
def poverty_socdem(cur_df):
    
    path = ('social_russia_data/poverty_socdem_2019.xls')
    df = pd.read_excel(path)
    name_val = list(df.iloc[1])
    name_val[0] = 'Район'
    
    df.columns = name_val
    df.drop(labels = [0, 1],axis = 0, inplace = True)
    df['Район'] = df['Район'].apply(lambda x: x.strip())
    df = predobr(df, 'Район')
    df.loc[df['Район'] == 'Тюменская область (кроме Ханты-Мансийского автономного округа-Югры и Ямало-Ненецкого автономного округа)', 'Район'] = 'Тюменская область безавтономных округов'
    df.loc[df['Район'] == 'Архангельская область (кроме Ненецкого автономного округа)', 'Район'] = 'Архангельская область безавтономного округа'
    
    region_cur = set(cur_df['Район'])
    region_1  = set(df['Район'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете poverty = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))
        
    cur_df['socdem_<16_age'] = ''
    cur_df['socdem_<more_employable_age'] = ''
    cur_df['socdem_<employable_age'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'socdem_<16_age'] = list(df[df['Район'] == el]['Дети в возрасте до 16 лет'])[0]
        cur_df.loc[cur_df['Район'] == el, 'socdem_<more_employable_age'] = list(df[df['Район'] == el]['Население старше трудоспособного возраста'])[0]
        cur_df.loc[cur_df['Район'] == el, 'socdem_<employable_age'] = list(df[df['Район'] == el]['Население трудоспособного возраста'])[0]
        
    
    return cur_df
```

```python
def housing(cur_df):
    path = ('social_russia_data/housing_2020.xlsx')
    df = pd.read_excel(path)

    for el in df['Регион']:
        if el.find('авт.') >= 0:
            el1 = el.replace("авт.", "автономный")
        else:
            el1 = el.replace("\n", "")
        if el.find("г.") >= 0:
            el1 = el.replace("г.", "город")
        if el.find(" - Югра") >= 0:
            el1 = el.replace(" - Югра", "")
            el1 = el1.replace("\n", "")
        df.loc[df['Регион'] == el, 'Регион'] = el1
    df.loc[df['Регион'] == 'городМосква', 'Регион'] = 'город Москва'
    df.loc[df['Регион'] == 'Республика Саха(Якутия)', 'Регион'] = 'Республика Саха (Якутия)'
    df.loc[df['Регион'] == 'Еврейская автономный область', 'Регион'] = 'Еврейская автономная область'
    df.loc[df['Регион'] == 'Архангельская область без автономный округа', 'Регион'] = 'Архангельская область безавтономного округа'
    df.loc[df['Регион'] == 'Bладимирская область', 'Регион'] = 'Владимирская область'
    df.loc[df['Регион'] == 'Bологодская область', 'Регион'] = 'Вологодская область'
    df.loc[df['Регион'] == 'Bоронежская область', 'Регион'] = 'Воронежская область'
    df.loc[df['Регион'] == 'Bолгоградская область', 'Регион'] = 'Волгоградская область'
    df.loc[df['Регион'] == 'Тюменская область без автономный округов', 'Регион'] = 'Тюменская область безавтономных округов'

    region_cur = set(cur_df['Район'])
    region_1  = set(df['Регион'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете cash_real_income = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))
        
    for el in region_cur:
        for column in list(df.columns[1:]):
            cur_df.loc[cur_df['Район'] == el, column] = list(df[df['Регион'] == el][column])[0]
    return cur_df
```

```python
def retail_turnover(cur_df):
    
    path = ('social_russia_data/retail_turnover_per_capita_2000_2021.xls')
    df = pd.read_excel(path)
    name_val = list(df.iloc[1])
    name_val[0] = 'Район'
    
    df.columns = name_val
    df.drop(labels = [0, 1, 2],axis = 0, inplace = True)
    
    df['Район'] = df['Район'].apply(lambda x: x.strip())
    df = predobr(df, 'Район')
    df.loc[df['Район'] == 'Тюменская область (кроме Ханты-Мансийского автономного округа-Югры и Ямало-Ненецкого автономного округа)', 'Район'] = 'Тюменская область безавтономных округов'
    df.loc[df['Район'] == 'Архангельская область (кроме Ненецкого автономного округа)', 'Район'] = 'Архангельская область безавтономного округа'
    region_cur = set(cur_df['Район'])
    region_1  = set(df['Район'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете poverty = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))

    cur_df['retail_turnover'] = ''
    for el in region_cur:
        cur_df.loc[cur_df['Район'] == el, 'retail_turnover'] = list(df[df['Район'] == el]['2019'])[0]
        
    return cur_df
```

```python
def crimes(cur_df):
    
    path = ('social_russia_data/crimes/4-EGS_Razdel_4_102019.xls')
    df = pd.read_excel(path)
    crimes = ['', 'Район', 'несовершеннолетними или при их соучастии', 'ранее совершавшими преступления', 'ранее судимыми', 'группой лиц', 'группой лиц по предварительному сговору', 'организованной группой', 'преступным сообществом', 'в состоянии алкогольного опьянения', 'в состоянии наркотического опьянения', 'в состоянии токсического опьянения' ]
    df.columns = crimes
    df = predobr(df, 'Район')
    df = df[11:]
    for el in df['Район']:
        if el.find("Республика") >= 0:
            el1 = el.replace("\n", " ")
        else:
            el1 = el.replace("\n", "")
        if el.find("г.") >= 0:
            el1 = el.replace("г.", "город")
        if el.find(" –  Югра") >= 0:
            el1 = el.replace(" –  Югра", "")
        if el.find(" – Кузбасс") >= 0:
            el1 = el.replace( " – Кузбасс", "")
        if el.find(" – Алания") >= 0:
            el1 = el.replace( " – Алания", "-Алания")
        if el.find("Новгородская  область") >= 0:
            el1 = el.replace("Новгородская  область","Новгородская область")
        df.loc[df['Район'] == el, 'Район'] = el1
    for el in list(df.columns[2:]):
        cur_df[el] = ''
        
    region_cur = set(cur_df['Район'])
    region_1  = set(df['Район'])
    region_cur = region_cur.intersection(region_1)
    print("кол-во регионов в датасете crimes = ", len(region_1))
    print("кол-во регионов poverty отличных от исходного = ", len(region_1) - len(region_cur))
    if len(region_1) - len(region_cur) > 0:
        print("Различия в регионе", region_1.difference(region_cur))
        print(set(cur_df['Район']).difference(region_cur))
    
    for region in region_cur:
        for column in list(df.columns[2:]):
            cur_df.loc[cur_df['Район'] == region, column] = list(df[df['Район'] == region][column])[0]
            
    return cur_df
    
```

## Оптимальное количество кластеров методом Локтя и методом Силуэта.

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

По методу локтя оптимальное количество кластеров - 3. Видно что SSE до 3 быстро начинает убывать(суммарное расстояние от центров до объектов кластера), после 3 начинает медленнее убывать. 

Также посчитаем оптимальное количество кластеров коэффициентом силуэта. Самое большое значение силуэта отвечает за самое оптимальное количество кластеров.

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Самое большое значение коэффициента 3 - это говорит о том, что это оптимальное количество кластеров.

```python
#подберем оптимальное количество кластеров для ЕМ-алгоритма с помощью коэф-та силуэта
silhouettes = []
clusters = range(2, 9)
for c in clusters:
    gm=GaussianMixture(n_components=c, random_state=42)
    c_pred=gm.fit_predict(scaled_features)
    silhouettes.append((c, metrics.silhouette_score(scaled_features, c_pred)))
silhouette_df=pd.DataFrame(silhouettes, columns=['Cluster', 'Sil_Score']).sort_values(by='Sil_Score', ascending=False)
silhouette_df
```

С помощью EM алгоритма оптимальное количество кластеров - 2. Для 3 значение примерно такое же как и для двух.
